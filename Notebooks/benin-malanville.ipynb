```python
Q1 = Data_benin.quantile(0.25)
Q3 = Data_benin.quantile(0.75)
IQR = Q3 - Q1

# Outliers are values outside the range [Q1 - 1.5*IQR, Q3 + 1.5*IQR]
outliers = ((Data_benin < (Q1 - 1.5 * IQR)) | (Data_benin > (Q3 + 1.5 * IQR)))

print("\nOutliers Detected:\n", outliers.sum())
```

```python
# Checking for negative values in GHI, DNI, DHI (should not have negatives)
incorrect_entries = (Data_benin[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']] < 0).sum()
print("\nIncorrect Entries (Negative Values):\n", incorrect_entries)
# Example of flagging suspicious entries in GHI
suspicious_entries_ghi = Data_benin[(Data_benin['GHI'] > 1400) | (Data_benin['GHI'] < 0)]  # GHI generally should be between 0 and 1400 W/m²
print("\nSuspicious GHI Entries:\n", suspicious_entries_ghi)
```

```python
# Replace negative values with 0
columns_to_correct = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']
Data_benin[columns_to_correct] = Data_benin[columns_to_correct].applymap(lambda x: max(x, 0))
```

```python
# Checking for negative values in GHI, DNI, DHI (should not have negatives)
incorrect_entries = (Data_benin[['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']] < 0).sum()
print("\nIncorrect Entries (Negative Values):\n", incorrect_entries)
```

# if we have null value

```python
threshold = len(Data_benin) * 0.5
Data_benin = Data_benin.dropna(axis=1,thresh = threshold)
```

```python
# Convert the Timestamp column to datetime format
Data_benin['Timestamp'] = pd.to_datetime(Data_benin['Timestamp'])
```

```python
# Plot line graphs for GHI, DNI, DHI, and Tamb
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.plot(Data_benin['Timestamp'], Data_benin['GHI'], label='GHI')
plt.title('Global Horizontal Irradiance over Time')
plt.xlabel('Timestamp')
plt.ylabel('GHI')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(Data_benin['Timestamp'], Data_benin['DNI'], label='DNI', color='orange')
plt.title('Direct Normal Irradiance over Time')
plt.xlabel('Timestamp')
plt.ylabel('DNI')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(Data_benin['Timestamp'], Data_benin['DHI'], label='DHI', color='green')
plt.title('Diffuse Horizontal Irradiance over Time')
plt.xlabel('Timestamp')
plt.ylabel('DHI')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(Data_benin['Timestamp'], Data_benin['Tamb'], label='Tamb', color='red')
plt.title('Tamprature over Time')
plt.xlabel('Timestamp')
plt.ylabel('Tamprature (°C)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Plot ModA and ModB with Cleaning events highlighted
plt.figure(figsize=(14, 8))

plt.plot(Data_benin['Timestamp'], Data_benin['ModA'], label='ModA', color='blue')
plt.plot(Data_benin['Timestamp'], Data_benin['ModB'], label='ModB', color='green')

# Highlight cleaning events
cleaning_times = Data_benin[Data_benin['Cleaning'] == 1]['Timestamp']
for ct in cleaning_times:
    plt.axvline(x=ct, color='red', linestyle='--', linewidth=1, alpha=0.6, label='Cleaning')

plt.title('Impact of Cleaning on ModA and ModB Sensor Readings')
plt.xlabel('Timestamp')
plt.ylabel('Sensor Readings')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Extract month and year for grouping
Data_benin['YearMonth'] = Data_benin['Timestamp'].dt.to_period('M')

# Group by YearMonth and calculate mean for each group
monthly_means = Data_benin.groupby('YearMonth').mean()

# Plot the monthly means
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
monthly_means['GHI'].plot(kind='line', marker='o')
plt.title('Average GHI by Month')
plt.xlabel('Month')
plt.ylabel('Average GHI')
plt.grid(True)

plt.subplot(2, 2, 2)
monthly_means['DNI'].plot(kind='line', marker='o', color='orange')
plt.title('Average DNI by Month')
plt.xlabel('Month')
plt.ylabel('Average DNI')
plt.grid(True)

plt.subplot(2, 2, 3)
monthly_means['DHI'].plot(kind='line', marker='o', color='green')
plt.title('Average DHI by Month')
plt.xlabel('Month')
plt.ylabel('Average DHI')
plt.grid(True)

plt.subplot(2, 2, 4)
monthly_means['Tamb'].plot(kind='line', marker='o', color='red')
plt.title('Average Tamb by Month')
plt.xlabel('Month')
plt.ylabel('Average Tamb (°C)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Boxplot for GHI to visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=Data_benin['GHI'])
plt.title('GHI Boxplot for Outlier Detection')
plt.show()
```

```python
# Select the relevant columns for the correlation matrix
corr_columns = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB']

# Calculate the correlation matrix
corr_matrix = Data_benin[corr_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix for Solar Radiation and Temperature Measures')
plt.show()
```

```python
# Plot pair plot
sns.pairplot(Data_benin[corr_columns], kind='reg', diag_kind='kde', plot_kws={'line_kws':{'color':'red'}})
plt.suptitle('Pair Plot for Solar Radiation and Temperature Measures', y=1.02)
plt.show()
```
